```python
import numpy as np
import pandas as pd
```

```python
!ls ../データ分析用ライブラリ練習/titanic
```

```python
train = pd.read_csv('../データ分析用ライブラリ練習/titanic/train.csv')
test = pd.read_csv('../データ分析用ライブラリ練習/titanic/test.csv')
gender_submission = pd.read_csv('../データ分析用ライブラリ練習/titanic/gender_submission.csv')
```

```python
gender_submission.head()
```

```python
train.head()
```

```python
data = pd.concat([train, test], sort=False)
data.isnull().sum()
```

```python
data['Sex'].replace(['male', 'female'], [0, 1], inplace = True)
```

```python
data['Embarked'].fillna(('S'), inplace=True)
data['Embarked'] = data['Embarked'].map({'S':0, 'C':1, 'Q':2}).astype(int)
```

```python
data['Fare'].fillna(np.mean(data['Fare']), inplace=True)
```

```python
age_avg = data['Age'].mean()
data['Age'].fillna(age_avg, inplace = True)
```

```python
data.drop(['Name', 'PassengerId', 'SibSp', 'Parch', 'Ticket', 'Cabin'], axis = 1, inplace = True)
```

```python
train = data[:len(train)]
test = data[len(train):]
```

```python
y_train = train['Survived']
X_train = train.drop('Survived', axis = 1)
X_test = test.drop('Survived', axis = 1)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(penalty = 'l2', solver = 'sag', random_state = 0)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred[:10]
```

```python
sub = pd.read_csv('../データ分析用ライブラリ練習/titanic/gender_submission.csv')
sub['Survived'] = list(map(int, y_pred))
sub.to_csv('submission.csv', index=False)
```

```python
train = pd.read_csv('../データ分析用ライブラリ練習/titanic/train.csv')
test = pd.read_csv('../データ分析用ライブラリ練習/titanic/test.csv')
gender_submission = pd.read_csv('../データ分析用ライブラリ練習/titanic/gender_submission.csv')
```

```python
data = pd.concat([train, test], sort=False)

data['Sex'].replace(['male', 'female'], [0, 1], inplace=True)
data['Embarked'].fillna(('S'), inplace=True)
data['Embarked'] = data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2}).astype(int)
data['Fare'].fillna(np.mean(data['Fare']), inplace=True)
data['Age'].fillna(data['Age'].median(), inplace=True)
```

```python
pd.plotting.scatter_matrix(data, figsize = (15,15))
```

```python
data['FamilySize'] = data['Parch'] + data['SibSp'] + 1
train['FamilySize'] = data['FamilySize'][:len(train)]
test['Familysize'] = data['FamilySize'][len(train):]
sns.countplot(x = 'FamilySize', data = train, hue = 'Survived')
```

```python
data['IsAlone'] = 0
data.loc[data['FamilySize'] == 1, 'IsAlone'] = 1
train['IsAlone'] = data['IsAlone'][:len(train)]
test['IsAlone'] = data['IsAlone'][len(train):]
```

```python
delete_columns = ['Name', 'PassengerId', 'SibSp', 'Parch', 'Ticket', 'Cabin']
data.drop(delete_columns, axis = 1, inplace = True)
train = data[:len(train)]
test = data[len(train):]

y_train = train['Survived']
X_train = train.drop('Survived', axis=1)
X_test = test.drop('Survived', axis=1)
```
