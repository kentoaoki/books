## 基本のネットワーク構造
- 同じ形状(ここではCNN)を2つのインプットで共有する
- 重みも共有し,同じように更新される
- アウトプットに対して,距離を計算し同じ画像はより近く,異なる画像はより遠くにする
    - 今回は距離の計算にはユークリッド距離,損失関数にはcontrastive loss function(類似度を距離で表示)を使用する
- 最終の出力は距離なので,例えば距離が最も近いものを同じラベルとする,とすれば分類にも使用できる




## Loss function

- Yが正解ラベル（1:Positive, 0:Negative）
- Dが予測距離
- 直感的には,Y=1の場合は第一項だけ残るためDが小さい方が、Y=0の場合は第二項だけ残りDが大きい方がLossが小さくなる

## 初期設定

```python
# import

import os
from tensorflow.keras.datasets import mnist
import tensorflow as tf
import numpy as np
import cv2
```

```python
# initialize

IMG_SHAPE = (28, 28, 1)
BATCH_SIZE = 64
EPOCHS = 10

BASE_OUTPUT = 'output'
MODEL_PATH = os.path.sep.join([BASE_OUTPUT, 'siamese_model'])
PLOT_PATH = os.path.sep.join([BASE_OUTPUT, 'plot.png'])
```

```python

import tensorflow.keras.backend as K
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ユークリッド距離を計算する関数
# レイヤとして埋め込むためにkerasで関数を生成する

def euclidean_distance(vectors):
    (A, B) = vectors
    sumSquared = K.sum(K.square(A - B), axis=1, keepdims=True)
    return K.sqrt(K.maximum(sumSquared, K.epsilon()))
```

```python
# 学習履歴のプロット、保存関数

def plot_training(history, plot_path):
    plt.style.use('ggplot')
    plt.figure()
    plt.plot(history.history['loss'], label='train loss')
    plt.plot(history.history['val_loss'], label='val loss')
    plt.plot(history.history['dist_accuracy'], label='train acc')
    plt.plot(history.history['val_dist_accuracy'], label='val_acc')
    plt.legend(loc = 'lower left')
    plt.savefig(plot_path)
```

## ペアデータの生成

ポジティブ、ネガティブの画像ペアを生成します

```python
def make_pairs(images, labels):
    pairImages = []
    pairLabels = []
    numClasses = len(np.unique(labels))
    idx = [np.where(labels == i)[0] for i in range(0, numClasses)]
    
    for idx_all in range(len(images)):
        current_img = images[idx_all]
        label = labels[idx_all]
        
        # possitive pair
        idx_b = np.random.choice(idx[label])
        pos_img = images[idx_b]
        pairImages.append([current_img, pos_img])
        pairLabels.append([1])
        
        # negative pair
        neg_idx = np.where(labels != label)[0]
        neg_img = images[np.random.choice(neg_idx)]
        pairImages.append([current_img, neg_img])
        pairLabels.append([0])
    
    return (np.array(pairImages), np.array(pairLabels))
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 正規化
x_train = x_train/255.0
x_test = x_test/255.0

# チャンネル用のダミー次元
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

(pairTrain, labelTrain) = make_pairs(x_train, y_train)
(pairTest, labelTest) = make_pairs(x_test, y_test)
```

```python
print(pairTrain.shape, labelTrain.shape)
print(pairTest.shape, labelTest.shape)
```

## モデルの作成

- ２つのミラーモデルを生成し、
- その２つのネトワークからの出力間の距離を計算する。それによって類似度を判定
- モデル全体の入力が２つあるのでfunctional API形式で記述する

```python
# import for model

from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, Lambda, Conv2D, Dense, Dropout,GlobalAveragePooling2D, MaxPool2D
```

```python

def build_siamese_model(input_shape, output_dim=48):
    inputs = Input(input_shape)
    x = Conv2D(64, (2,2), padding='same', activation='relu')(inputs)
    x = MaxPool2D(pool_size=(2,2))(x)
    x = Dropout(0.3)(x)

    x = Conv2D(64, (2,2), padding='same', activation='relu')(x)
    x = MaxPool2D(pool_size=2)(x)
    x = Dropout(0.3)(x)

    pooled_output = GlobalAveragePooling2D()(x)
    outputs = Dense(output_dim)(pooled_output)
    
    model = Model(inputs, outputs)
    return model

model = build_siamese_model((28,28,1))
model.summary()
```

```python

# １つのインスタンスを共有することでパラメータを共有する

imgA = Input(shape=IMG_SHAPE)
imgB = Input(shape=IMG_SHAPE)

feature_extractor = build_siamese_model(IMG_SHAPE)
modelA = feature_extractor(imgA)
modelB = feature_extractor(imgB)
```

```python

dist = Lambda(euclidean_distance)([modelA, modelB])
outputs = Dense(1, activation='sigmoid')(dist)
model = Model(inputs=[imgA, imgB], outputs=outputs)
```

```python
model.summary()
```

```python
def contrastive_loss(y, preds, margin=1):
    y = tf.cast(y, preds.dtype)
    squaredPreds = K.square(preds)
    squaredMargin = K.square(K.maximum(margin - preds, 0))
    loss = K.mean(y*squaredPreds + (1 - y)*squaredMargin)
    return loss
```

```python
# 0に近い方場合同一として判定

def dist_accuracy(y, preds):
    return K.mean(K.equal(y, K.cast(preds < 0.5, y.dtype)))
```

```python
# binary_crossentropyの方が簡単に試せる
model.compile(
    loss = contrastive_loss,
    optimizer = 'adam',
    metrics = [dist_accuracy]
)
```

```python
history = model.fit(
    [pairTrain[:,0], pairTrain[:,1]], labelTrain[:],
    validation_data=([pairTest[:,0], pairTest[:,1]], labelTest[:]),
    batch_size = BATCH_SIZE,
    epochs = EPOCHS
)
```

```python

model.save(MODEL_PATH)
plot_training(history, PLOT_PATH)
```
