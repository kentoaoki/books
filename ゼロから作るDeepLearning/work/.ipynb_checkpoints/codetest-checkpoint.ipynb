```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1

AND(0,0), AND(0,1), AND(1,0), AND(1,1)
```

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

AND(0,0), AND(0,1), AND(1,0), AND(1,1)
```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

for i in range(2):
    for j in range(2):
        print(f'if {i} and {j}:', NAND(i, j), OR(i, j))
```

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    return AND(s1, s2)

for i in range(2):
    for j in range(2):
        print(f'if {i} and {j}:', XOR(i, j))
```

```python
def step_function(x):
    y = x > 0
    return y.astype(np.int)

x = np.arange(-5,5,0.1)
y = step_function(x)
plt.plot(x, y)
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

x = np.arange(-5,5,0.1)
y = sigmoid(x)
plt.plot(x,y)
```
