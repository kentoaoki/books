```python
class UnionFind:
    def __init__(self, n):
        self.par = [i for i in range(n+1)]
        self.rank = [0] * (n+1)
        self.size = [1] * (n+1) # 親に集約される

    # 検索
    def find(self, x):
        if self.par[x] == x:
            return x
        else:
            self.par[x] = self.find(self.par[x])
            return self.par[x]

    # 併合
    def union(self, x, y):
        x = self.find(x)
        y = self.find(y)

        if x == y:
            return

        if self.rank[x] < self.rank[y]:
            self.par[x] = y
            self.size[y] += self.size[x]
            self.size[x] = 0
        else:
            self.par[y] = x
            self.size[x] += self.size[y]
            self.size[y] = 0
            if self.rank[x] == self.rank[y]:
                self.rank[x] += 1

    # 同じ集合に属するか判定
    def same(self, x, y):
        return self.find(x) == self.find(y)

    # すべての頂点に対して親を検索する
    def all_find(self):
        for n in range(len(self.par)):
            self.find(n)
```

```python
# code 13.1

def search(G,s):
    """
    グラフGに対し、頂点sを始点とする探索
    """
    N = len(G)
    todo = deque() # BFSならキュー、DFSならスタックとして使用
    seen = [False]*N # 発見チェック（seen:True & not in todo => 訪問済み)
    
    # 初期状態
    seen[s] = True
    todo.append(s)
    
    # todoが空になるまで探索を行う
    while todo:
        v = todo.popleft() # DFSならpop()
        for x in G[v]:
            if seen[x]: # 発見済み
                continue
            seen[x] = True
            todo.append(x)
```

```python
# code 13.3

from collections import deque

def BFS(G, s):
    """
    グラフGに関して,頂点sからの最短路長を求める
    input: グラフG, 探索の視点s
    data: sからの距離dist配列(!=-1なら発見済み), 探索予定をキューとして格納するque
    output: sから各頂点への最短路長を表すdist配列
    """
    # 初期設定
    N = len(G)
    dist = [-1]*N
    que = deque()
    dist[s] = 0
    que.append(s)
    
    # 探索
    while que:
        v = que.popleft()
        for x in G[v]:
            if dist[x] != -1:
                continue
            dist[x] = dist[v] + 1
            que.append(x)
    return dist

N,M = map(int, input().split())
G = [[] for i in range(N)]
for i in range(M):
    a,b = map(int, input().split())
    a -= 1
    b -= 1
    G[a].append(b)

dist = BFS(G,0)
print(dist)
```

```python
# code 13.5
"""
二部グラフの判定。seenを0,1に塗り分ける
"""

def dfs(G, v, cur):
    """
    頂点vから辿れる頂点を全て訪問する（再帰）
    """
    # 訪問時の処理
    seen[v] = cur
    
    # 再帰的に探索
    for next_v in G[v]:
        if seen[next_v] != -1:
            if seen[next_v] == cur: # 隣接頂点が同じ色の時
                return False
            continue
        if not dfs(G, next_v, 1-cur):
            return False
    
    # 帰りがけの処理
    return True
        
N,M = map(int,input().split())
G = [[] for i in range(N)] # 隣接リスト
for i in range(M):
    a,b = map(int, input().split())
    a -= 1
    b -= 1
    G[a].append(b)

seen = [-1]*N
ans = True
for v in range(N):
    if seen[v] != -1:
        continue
    if not dfs(G,v,0):
        ans = False
print('Yes' if ans else 'No')
```

```python
# code 13.6

def topological(G, v):
    """dfsを用いて、トポロジカルソートを作る
    
    Args:
        G: 探索するグラフ(隣接リスト)
        v: 探索開始頂点
        
    Returns:
        -
        (order: v頂点からのトポロジカルソートが逆順に格納される)
        
    Notes:
        orderには帰りがけ順に格納される
    
    """
    seen[v] = True
    for next_v in G[v]:
        if seen[next_v]:
            continue
        topological(G, next_v)
    order.append(v)

N,M = map(int, input().split())
G = [[] for i in range(N)]
for i in range(M):
    a,b = map(int, input().split())
    G[a].append(b)
    
seen = [False]*N
order = []
for v in range(N):
    if seen[v]:
        continue
    topological(G, v)
order = list(reversed(order))
print(order)
```

```python
# code 13.9

def dfs(G, v, p, d):
    """vを根とする木について行きがけ/帰りがけを利用し、各頂点の深さと部分木サイズを求める
    
    Args:
        G: グラフ(隣接リスト)
        v: 探索開始頂点
        p: 親頂点
        d: 頂点vの深さ
    
    Returns:
        depth: 各頂点の根からの深さ
        subtree_size: 各頂点からの部分木のサイズ
        
    Notes:
        リストdepth,subtree_sizeを関数外で作成して使用する
    """
    # 行きがけ処理
    depth[v] = d
    # 探索
    for c in G[v]:
        if c == p: # 親への逆流を防ぐ
            continue
        dfs(G, c, v, d+1)
    # 帰りがけ処理
    subtree_size[v] = 1
    for c in G[v]:
        if c == p:
            continue
        subtree_size[v] += subtree_size[c]
        
N = int(input())
G = [[] for i in range(N)]
for i in range(N-1):
    a,b = map(int, input().split())
    G[a].append(b)
    G[b].append(a)

root = 0
depth = [0]*N
subtree_size = [0]*N
dfs(G,root,root,0)
print(depth,subtree_size)
```

```python
# code 14.2

N,M,s = map(int, input().split())
G = [[] for i in range(N)]
for i in range(M):
    a,b,w = map(int, input().split())
    G[a].append([b,w])

def bellman_ford(G, s):
    """ベルマンフォード法により,負の重みも持つグラフの最短路を求める O(|V|*|E|)
    
    Args:
        G: 重み付きグラフ([頂点,重み]で隣接リスト管理)
        s: 探索開始頂点
        
    Returns:
        if 負閉路を持つ:
            -1: 負閉路をもつフラグとして返す
        else:
            dist: 探索開始頂点からの最短距離を表す
    
    Notes:
    
    """
    # 前処理
    INF = 10**7
    N = len(G)
    exist_negative_cycle = False
    dist = [INF]*N
    dist[s] = 0
    
    # N回探索する
    for iter in range(N):
        update = False
        for v in range(N):
            if dist[v] == INF:
                continue
            for e in G[v]:
                if dist[e[0]] > dist[v] + e[1]:
                    dist[e[0]] = dist[v] + e[1]
                    update = True
        if not update:
            break
        if iter == N-1 and update: # N回目の反復で更新が行われれば負閉路を持つ
            exist_negative_cycle = True
    
    # Return
    if exist_negative_cycle:
        return -1
    else:
        return dist
        
ans = bellman_ford(G,s)
print(ans)
```

```python
# code 14.3

def dijkstra(G,s):
    """ダイクストラ法により、辺の重みが非負グラフの最短経路を求める O(|V|^2)
    
    Args:
        G: 重み付きグラフ((頂点,重み)の隣接リスト形式)
        s: 探索開始頂点
    
    Returns:
        dist: s(探索開始頂点)からの最短距離.INFの場合は到達しない
        
    """
    INF = float('inf')
    used = [False]*N
    dist = [INF]*N
    dist[s] = 0
    for iter in range(N):
        # 未使用かつdist値最小の頂点を探す
        min_dist = INF
        min_v = -1
        for v in range(N):
            if not used[v] and dist[v] < min_dist:
                min_dist = dist[v]
                min_v = v
        if min_v == -1:
            break
        for v,w in G[min_v]:
            dist[v] = min(dist[v], dist[min_v] + w)
        used[min_v] = True
    return dist

N,M,s = map(int, input().split())
G = [[] for i in range(N)]
for i in range(M):
    a,b,w = map(int, input().split())
    G[a].append([b,w])
ans = dijkstra(G,s)
ans
```

```python
# code 14.5

INF = 10**7
N,M = map(int, input().split())
dp = [[INF]*N for i in range(N)]
for i in range(M):
    a,b,w = map(int, input().split())
    dp[a][b] = w

for v in range(N):
    dp[v][v] = 0

def worshall_floyd(G):
    """ワーシャルフロイド法により、グラフの全点間最短距離を返す
    
    Args:
        G: グラフ(隣接行列,距離不明辺はINFで初期化)
    
    Returns:
        G: 最短距離の記述されたグラフ(INFは到達不能,マイナスは負閉路)
    """
    # 自身への距離をゼロで初期化
    N = len(G)
    for i in range(N):
        G[i][i] = 0
    # 探索
    for k in range(N):
        for i in range(N):
            for j in range(N):
                G[i][j] = min(G[i][j], G[i][k] + G[k][j])
    return G
ans = worshall_floyd(dp)
print(ans)
```

```python
class UnionFind:
    """Union-Find法を活用し、データのグループ分けを行う
    
    Attributes:
        par: 親のインデクスをさす
        rank: グループの高さ
        size: グループのサイズ(親に集約される)
        
    Args:
        n: グループ分けする要素の個数
    
    """
    def __init__(self, n):
        self.par = [i for i in range(n+1)]
        self.rank = [0] * (n+1)
        self.size = [1] * (n+1) # 親に集約される

    # 検索
    def find(self, x):
        if self.par[x] == x:
            return x
        else:
            self.par[x] = self.find(self.par[x])
            return self.par[x]

    # 併合
    def union(self, x, y):
        x = self.find(x)
        y = self.find(y)

        if x == y:
            return

        if self.rank[x] < self.rank[y]:
            self.par[x] = y
            self.size[y] += self.size[x]
            self.size[x] = 0
        else:
            self.par[y] = x
            self.size[x] += self.size[y]
            self.size[y] = 0
            if self.rank[x] == self.rank[y]:
                self.rank[x] += 1

    # 同じ集合に属するか判定
    def same(self, x, y):
        return self.find(x) == self.find(y)

    # すべての頂点に対して親を検索する
    def all_find(self):
        for n in range(len(self.par)):
            self.find(n)
```

```python
# code 15.1

N,M = map(int, input().split())
edge = []
for i in range(M):
    u,v,w = map(int, input().split())
    edge.append([w,[u,v]])

edge.sort()
res = 0
uf = UnionFind(N)
for i in range(M):
    w = edge[i][0]
    u = edge[i][1][0]
    v = edge[i][1][1]
    if uf.same(u,v):
        continue
    res += w
    uf.union(u,v)
print(res)
```

```python
def rec(i,w,a):
    if i == 0:
        if w == 0:
            return True
        else:
            return False
    if rec(i-1,w,a) or rec(i-1,w-a[i-1],a):
        return True
    return False

n,w = map(int, input().split())
a = list(map(int, input().split()))
if rec(n,w,a):
    print('Yes')
```
